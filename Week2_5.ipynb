```python
def degerlendir(row):
    actor = row.iloc[1]
    budget = row.iloc[4]

    if actor == "Pierce Brosnan":
        return "En iyi film"
    elif actor == "Roger Moore" and budget > 40:
        return "Eglenceli film"
    else:
        return "Fikrim Yok"
```
